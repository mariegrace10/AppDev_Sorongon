1.

    A. Class: An object's blueprint is called a class. It outlines the methods (functions) and characteristics that the class's objects will possess. To put it another way, a class is a prototype or template that may be used to create objects with comparable characteristics and actions. An object is a specific instance of a class. Based on a class, it is a concrete entity with distinct attributes, behavior, and identity (methods). Classes are used to build objects.An instance is a particular realization of an object. It speaks of a specific item made from a class. A class's instances all share the same behavior, but they also each have their own distinct state.

    B. A method is a function that may be invoked on instances of a class that it is connected with. Methods work with the data that is present in class instances. A function is an independent function can be called on any class or object, and it is a block of code that carries out a particular task.

    C. An attribute is a piece of information associated with an object or class. A class's or its instance's attributes can be variables that are accessible using dot notation. A designated storage place that contains a value is called a variable. When discussing classes and objects, variables frequently allude to the traits or properties that are connected to them.

2. A Procedural Paradigm is writing functions or procedures that manipulate data is the main goal of procedural programming. Programs are set up as a sequence of sequentially executed stages or procedures. Functions and data are kept apart, with data being transferred between them as required. On an
OOP Paradigm, rather than basing software design on actions and logic, object-oriented programming, or OOP, focuses on structuring data and objects. The idea of objects with data (attributes) and activity (methods) that interact with one another is emphasized. Encapsulation, inheritance, and polymorphism are made possible by OOP, resulting in code that is more modular, reusable, and manageable.

3. In object-oriented programming, there are benefits to using classes as opposed to modules that only hold data and functions. Classes encourage modularity and simplify systems by combining data and behavior into a single unit. They allow for more control over access to class internals, flexibility, and code reuse by supporting features like inheritance, polymorphism, and data hiding. Additionally, classes help with abstraction by simplifying the modeling of complicated systems and improving the readability and maintainability of the code. All things considered, taking classes offers a more methodical and systematic approach to programming, especially when creating intricate systems with several interdependent parts.
