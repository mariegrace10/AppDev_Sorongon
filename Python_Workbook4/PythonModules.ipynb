- What are the constants in the math module?
- What's does `math.exp` do? What about `math.pow`?  
- How do you calculate logarithms? Logarithms base 10? Natural Logarithms?


A.  There are a number of mathematical constants in the Python math module, such as math.pi, math.e, math.tau, math.inf, math.nan.

B. math.exp is the function that takes a parameter (e, which is approximately equivalent to 2.71828) and returns e raised to the power of x. Stated differently, e^x is computed by math.exp(x).
The value of x raised to the power of y is returned by the math.pow function. It is comparable to Python's ** operator, but it always yields a floating-point output. As an illustration, math.pow(2, 3) yields 8.0.

C. Base 10 logarithms (log10): The math.log10(x) function can be used to get the logarithm of a given number x. For instance, since 10^2 equals 100, math.log10(100) yields 2.0.

Natural logarithms (ln or logarithms base e): The math.log(x) function can be used to get the natural logarithm of a number x, or the logarithm to the base e. For instance, because e^1 equals e, math.log(math.e) returns 1.0.

Custom base logarithms: The change of base formula, log_b(x) = log(x) / log(b), can be used to compute logarithms with a custom base b. It would be expressed as math.log(x) / math.log(b) in Python.

```python
%%writefile gradesCalculation.py

def calculate_average(grades):

    return sum(grades) / len(grades)

def determine_status(average_grade):

    if average_grade < 75:
        return 'Not Passed'
    else:
        return 'Passed'
```

```python
import gradesCalculation

math_grades = [92, 93]
science_grades = [93, 94]
english_grades = [94, 93]

math_average = gradesCalculation.calculate_average(math_grades)
science_average = gradesCalculation.calculate_average(science_grades)
english_average = gradesCalculation.calculate_average(english_grades)

math_status = gradesCalculation.determine_status(math_average)
science_status = gradesCalculation.determine_status(science_average)
english_status = gradesCalculation.determine_status(english_average)

print("Student Status:", math_average, math_status)
print("Student Status:", science_average, science_status)
print("Student Status:", english_average, english_status)
```
