#### Exercise 1

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E4.6

```python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

def computePay(hours, rate):

    if hours < 40:
        pay = hours * rate
    else:
        pay1 = 40*rate
        pay2 = (hours-40)*rate*1.5
        pay = pay1+pay2
    print("Pay:", pay)

    computePay(hours, rate)
```

#### Challenge

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

    Run the program repeatedly to test the various different values for input.
    
Adapted from [PEW] E4.6
<br>

```python
score = float(input("Enter score between 0 and 100: "))

def computegrade(score):

    if score >= 97:
        print("Grade: 1.0")
    elif score >= 93:
        print("Grade: 1.25")
    elif score >= 89:
        print("Grade: 1.5")
    elif score >= 85:
        print("Grade: 1.75")
    elif score >= 81:
        print("Grade: 2.0")
    elif score >= 77:
        print("Grade: 2.25")
    elif score >= 74:
        print("Grade: 2.5")
    elif score >= 0:
        print("Grade: 3.0")
    else:
        print("Invalid!")


computegrade(score)
```
