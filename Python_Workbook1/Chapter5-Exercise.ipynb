Investigate on the difference between `is` and `==` given the following variables.
```python
a = 2
b = 2.0
c = a
```

````python
"a is c" will return TRUE because a = 2 and c = a which also has the value or integer or has the same memory location.
````

````python
"a is b" will return FALSE because a = 2 and b = 2.0. In other words, they have different memory location.
````

````python
"a == b" will return TRUE because a = 2 and b = 2.0. is equal because they have the same value which is 2 despite a being an integer and being a float.
````

````python
"a == c" will return TRUE because a and c has the same value which is 2.
````

#### Challenge 1

Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

    Enter a number: 4
    Enter a number: 5
    Enter a number: bad data
    Invalid input
    Enter a number: 7
    Enter a number: done
    16 3 5.333333333333333
    
[PEW] E5.1

````python
class UsersInput:

    def __init__(self):
        self.count = 0
        self.total = 0



    def getCount(self):

        while True:
            inpt = input("Enter a number: ")

            if inpt == 'done':
                break

            elif inpt.isdigit():
                num = float(inpt)
                self.total += num
                self.count += 1

            else:
                print("bad data")
                print("Invalid input")

        return self.count



    def getTotal(self):

        if self.count == 0:
            print("No data")

        else:
            average = self.total / self.count
            print(self.total, self.count, average)

        return self.total



def main():

    ui = UsersInput()
    count = ui.getCount()
    if count != 0:
        ui.getTotal()


main()
````

#### Challenge 2  

Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

[PEW] E5.2

````python
class UsersInput:

    def __init__(self):
        self.count = 0
        self.total = 0
        self.numbers = []



    def getCount(self):

        while True:
            inpt = input("Enter a number: ")

            if inpt == 'done':
                break

            elif inpt.isdigit():
                num = float(inpt)
                self.total += num
                self.count += 1
                self.numbers.append(num)

            else:
                print("bad data")
                print("Invalid input")

        return self.count



    def getTotal(self):

        if self.count == 0:
            print("No data")

        else:
            print(self.total, self.count)

            max_num = max(self.numbers)
            min_num = min(self.numbers)
            print(max_num, min_num)

        return self.total


def main():

    ui = UsersInput()
    count = ui.getCount()
    if count != 0:
        ui.getTotal()


main()
````
